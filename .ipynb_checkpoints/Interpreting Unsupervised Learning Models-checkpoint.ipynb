#  Arctic Penguin Exploration: Unraveling Clusters in the Icy Domain with K-means clustering




**Origin of this data** : Data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network

**The dataset consists of 5 columns.**

- culmen_length_mm: culmen length (mm)
- culmen_depth_mm: culmen depth (mm)
- flipper_length_mm: flipper length (mm)
- body_mass_g: body mass (g)
- sex: penguin sex

Unfortunately, they have not been able to record the species of penguin, but they know that there are three species that are native to the region: **Adelie**, **Chinstrap**, and **Gentoo**, so our task is to identify the groups in the dataset!

# Data Processing

```python
# Import Required Packages
import warnings
import shap # Model explainability

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

warnings.filterwarnings("ignore")
```

# Prepare data

```python
# Loading and examining the dataset
penguins_df = pd.read_csv("penguins.csv")

penguins_df.head()
```

```python
penguins_df.info()
```

```python
# check null values

penguins_df.isna().sum().plot.barh()
plt.title("Null values");
```

```python
# remove null values
penguins_df.dropna(inplace=True)

penguins_df.isna().sum()
```

```python
penguins_df.describe()
```

```python
penguins_df.sex.value_counts()
```

```python
# remove invalid sex values

penguins_df.drop(penguins_df[penguins_df["sex"] == "."].index, inplace = True)

penguins_df["sex"].unique()
```

```python
# pair plot

sns.pairplot(penguins_df);
```

```python
# check correlation

sns.heatmap(penguins_df.corr(), annot=True);
```

```python
# boxplot
penguins_df.boxplot()
```

### remove outliers

1. Calculate the first and third quartiles of prices and store these as q1 and q3, respectively.
2. Calculate the IQR score into IQR.
3. Calculate the lower outlier limit into lower_limit.
4. Calculate the upper outlier limit into upper_limit.

```python
# Calculate the 25th and 75th percentiles to identify outliers
q1 = penguins_df["flipper_length_mm"].quantile(0.25)
q3 = penguins_df["flipper_length_mm"].quantile(0.75)

# Find the IQR (Interquartile range)
IQR = q3 - q1
factor = 1.5

# Calculate the lower limit
lower_limit = q1 - (IQR * factor)

# Calculate the upper limit
upper_limit = q3 + (IQR * factor)

penguins_clean = penguins_df[(penguins_df["flipper_length_mm"] < upper_limit) &
                             (penguins_df["flipper_length_mm"] > lower_limit)]

print(f"shape of penguins_clean: {penguins_clean.shape}")
```

```python
# confirm there is no outlier

penguins_clean.boxplot()
```

## Preprocessing


```python
# Create dummy variables for the available categorical feature in the dataset, then drop the original column.
df = pd.get_dummies(penguins_clean)

df.head()
```

```python
# scale the dataset

# instantiate StandardScaler
ss = StandardScaler()

# perform fit_transform on features
X = ss.fit_transform(df)

# put result into dataframe
df_processed = pd.DataFrame(X, columns=df.columns)
df_processed.head()
```

# Model Building

### Hyperparameter tuning

First, we will try to find out the optimal number of `k` parameter to use in this model.

```python
# Detect the optimal number of clusters for k-means clustering
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df_processed)
    inertia.append(kmeans.inertia_)    
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show();
```

We can see that after the k-value of `4`, the curve begins to flattern out, so our optimal k-value is 4.

### Build final model

```python
n_clusters=4

# Run the k-means clustering algorithm with the optimal number of clusters 

kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(df_processed)
```

# Communicate

### Perform PCA  

```python
# instantiate PCA with without specifying the number of components
pca = PCA(n_components=None)

# fit pca to the features
df_pca = pca.fit(df_processed)

# calculate explained variance ratio for the features
df_pca.explained_variance_ratio_

# check the number of features with more than 10% (0.1) explained variance ratio
n_components=sum(df_pca.explained_variance_ratio_>0.1)

print(f"Explained variance ration: {df_pca.explained_variance_ratio_}\n", f"n_components: {n_components}")
```

```python
# final PCA 
# instantiate PCA with without specifying the number of components
pca = PCA(n_components=n_components)

# fit pca to features
penguins_PCA = pca.fit_transform(df_processed)
```

```python
# visualize the resulting clusters

plt.scatter(penguins_PCA[:, 0], penguins_PCA[:, 1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title(f'K-means Clustering (K={n_clusters})')
plt.legend()
plt.show()
```

## Statiscal dataframe for each cluster

```python
penguins_clean['label'] = kmeans.labels_
numeric_columns = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g','label']
stat_penguins = penguins_clean[numeric_columns].groupby('label').mean()
stat_penguins
```

```python
# visualization

stat_penguins.plot.bar();
```

We can see that Penguins who fall into `cluster 2` have longer culmen and flipper and a larger body mass than the other penguins.

# Model Explainability

```python
# Create a proxy model that returns the cluster labels
def proxy_model(X):
    return kmeans.predict(X)
```

### Setting up SHAP Explainer
Now comes the model explainer part.
We will first create an explainer object by providing a KMeans model, then calculate SHAP value using the features.

```python
explainer = shap.Explainer(proxy_model,df_processed)

shap_values = explainer(df_processed)
```

#### Summary Plot
Display the summary_plot using SHAP values

```python
shap.summary_plot(shap_values, df_processed)
```

The summary plot shows the feature importance of each feature in the model. The results show that **`flipper length`** and **`body mass`** play major roles in determining the results.
